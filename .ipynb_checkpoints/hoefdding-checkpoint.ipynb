### Höffding

Olgu $a_i < X_i < b_i$, $S_n = \sum_{i=1}^{n} X_i$

\begin{equation*}
    P ( \mid S_n - ES_n \mid \geq c ) \leq 2 \exp{ \left( - \frac{ 2 c^2 }{ \sum_{i=1}^{n} (b_i - a_i)^2 } \right) }
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

%matplotlib inline
```

\begin{equation*}
    \varepsilon = \frac{c}{n} = \sqrt{ - \frac{ \ln{ \left( 1 - \alpha \right) } }{ 2 n } }
\end{equation*}

```python
n = np.arange(1, 100, 1)
alpha = 0.95
epsilon = (- np.log((1 - alpha) / 2) / (2 * n)) ** 0.5

plt.plot(n, epsilon)

plt.legend([f'α = {alpha}'])
plt.xlabel('n')
plt.ylabel('ɛ')
```

$X_i \in \{ 0, 1 \} \implies S_n \sim Bin(n, p)$

```python
p = 0.9
n = 200

p - binom.ppf(q=0.025, p=p, n=n) / n
```
