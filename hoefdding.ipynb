### Höffdingi võrratus

Olgu $a_i < X_i < b_i$ ja $S_n = \sum_{i=1}^{n} X_i$

\begin{equation*}
    P ( \mid S_n - ES_n \mid \geq c ) \leq 2 \exp{ \left( - \frac{ 2 c^2 }{ \sum_{i=1}^{n} (b_i - a_i)^2 } \right) }
\end{equation*}

Kui $X_i \in \{ 0, 1 \} \implies S_n \sim Bin(n, p)$
\begin{equation*}
    P ( \mid \frac{ S_n }{ n } - p \mid \geq \frac{ c }{ n } ) \leq 2 \exp{ \left( - \frac{ 2 c^2 }{ n } \right) }
\end{equation*}

$\alpha \in [0, 1]$, siis Höffdingi võrratusest

\begin{equation*}
     \varepsilon := \frac{c}{n} = \sqrt{ - \frac{1}{2n} \ln{ \left( \frac{ 1 - \alpha }{2} \right) } }
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

%matplotlib inline
```

```python
n = np.arange(1, 51)
alpha = np.array([0.5, 0.7, 0.95])

for a in alpha:
    epsilon = (- np.log((1 - a) / 2) / (2 * n)) ** 0.5
    plt.plot(n, epsilon)

plt.legend([f'α = {a}' for a in alpha])
plt.xlabel('n')
plt.ylabel('ɛ')
```

... $p$ kasutav hinnang ...

\begin{equation*}
    P ( S_n \leq np - n \varepsilon ) \leq \frac{ 0.05 }{ 2 }
\end{equation*}

```python
n = np.arange(1, 51)
ps = np.array([0.5, 0.7, 0.9])

for p in ps:
    plt.plot(n, p - binom.ppf(q=0.025, p=p, n=n) / n)
    
plt.legend([f'p = {p}' for p in ps])
plt.xlabel('n')
plt.ylabel('ɛ')
```
